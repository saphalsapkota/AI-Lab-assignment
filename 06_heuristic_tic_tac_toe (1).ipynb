## 6. WAP in Python to calculate the heuristic value of the states for Tic-Tac-Toe

**Heuristic function:**
e(p) = No. of complete rows, columns or diagonals are still open for player ) â€“ (No. of complete rows, columns or diagonals are still open for  opponent)

---

Here, heuristic function is evaluated for some test Tic-Tac-Toe states of size 3x3:

```python
def heuristic(state, player_X=True):
    """
    Calculate heuristic value for any particular tic-tac-toe state.
    """
    lines = [
        [0, 1, 2],
        [3, 4, 5],
        [6, 7, 8],
        [0, 3, 6],
        [1, 4, 7],
        [2, 5, 8],
        [0, 4, 8],
        [2, 4, 6],
    ]  # index numbering for rows, columns or diagonals (3x3)
    x_open = o_open = common = 0
    for line in lines:
        items = []
        for index in line:
            i = index // 3
            j = index % 3
            items.append(state[i][j])
        if "X" in items and "O" in items:
            continue
        elif "X" in items and "" in items:
            x_open += 1
        elif "O" in items and "" in items:
            o_open += 1
        elif "" in items:
            common += 1

    x_open += common
    o_open += common

    h = x_open - o_open if player_X else o_open - x_open

    return h


def print_state(state):
    length = len(state)
    for i, row in enumerate(state):
        print(" | ".join([i for i in row]))
        if i < length - 1:
            print("--  --  --")


test_1 = [
    ["X", "X", "O"],
    ["O", "", "O"],
    ["", "O", ""],
]

test_2 = [["X", "", "X"], ["", "", "X"], ["O", "", ""]]

test_3 = [
    ["X", "", "O"],
    ["O", "", "O"],
    ["O", "X", ""],
]

# Calculate heuristic values assuming player is `X` for various states
print_state(test_1)
print(f"\nHeuristic value (Test 1): {heuristic(test_1)}\n")
print_state(test_2)
print(f"\nHeuristic value (Test 2): {heuristic(test_2)}\n")
print_state(test_3)
print(f"\nHeuristic value (Test 3): {heuristic(test_3)}\n")
```
