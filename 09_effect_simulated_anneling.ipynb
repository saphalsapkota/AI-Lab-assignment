## 9. WAP to demonstrate the effect of temperature on the probability of choosing an inferior node by selecting an appropriate temperature schedule.

```python
import math
import matplotlib.pyplot as plt
import numpy as np

# Probability of accepting an inferior solution
def acceptance_probability(delta, temperature):
    if delta < 0:
        return 1.0  # Always accept better solutions
    else:
        return math.exp(-delta / temperature)

# Simulate the effect of temperature on acceptance probabilities
def simulate_temperature_effect():
    temperatures = np.logspace(0, -3, 100)  # Temperature schedule (from 1 to 0.001)
    delta = 5  # Energy difference (inferior node quality)

    probabilities = [acceptance_probability(delta, temp) for temp in temperatures]

    # Plotting the results
    plt.figure(figsize=(8, 6))
    plt.plot(temperatures, probabilities, label=f"Delta = {delta}")
    plt.xscale("log")
    plt.xlabel("Temperature (log scale)")
    plt.ylabel("Probability of Accepting Inferior Node")
    plt.title("Effect of Temperature on Acceptance Probability")
    plt.grid(True)
    plt.legend()
    plt.show()

simulate_temperature_effect()
```
